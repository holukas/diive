```python
import ipywidgets as widgets
from diive.configs.exampledata import load_exampledata_parquet
```

```python
df = load_exampledata_parquet()
df = df.head(1000).copy()
df.head(3)
```

```python
w = widgets.Dropdown(
    options=['1', '2', '3'],
    value='2',
    description='Number:',
    disabled=False,
)
display(w)
```

```python
def on_change(change):
    return w.value
    # df[var].plot();  
    # plt.show()
w.observe(on_change)
```
