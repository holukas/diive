# **Input data**

```python
df = exd.load_exampledata_parquet()
df
```

## Remove columns

```python
dropcols = [
    'E_0_CUT_REF',
    'FP_alpha',
    'FP_alpha_sd',
    'FP_beta',
    'FP_beta_sd',
    'FP_dRecPar',
    'FP_E0',
    'FP_E0_sd',
    'FP_errorcode',
    'FP_GPP2000',
    'FP_k',
    'FP_k_sd',
    'FP_NEEnight_CUT_REF',
    'FP_qc',
    'FP_RRef',
    'FP_RRef_Night',
    'FP_RRef_sd',
    'FP_Temp_CUT_REF',
    'FP_VARday',
    'FP_VARnight',
    'Rg_fall',
    'Rg_fall_qc',
    'Rg_fmeth',
    'Rg_fnum',
    'Rg_fqc',
    'Rg_fsd',
    'Rg_fwin',
    'Tair_fall',
    'Tair_fall_qc',
    'Tair_fmeth',
    'Tair_fnum',
    'Tair_fqc',
    'Tair_fsd',
    'Tair_fwin',
    'Unnamed: 0',
    'VPD_fall',
    'VPD_fall_qc',
    'VPD_fmeth',
    'VPD_fnum',
    'VPD_fqc',
    'VPD_fsd',
    'VPD_fwin',
    'NEW_FP_Temp',
    'NEW_FP_VPD',
    'LE_fall',
    'LE_fall_qc',
    'LE_fmeth',
    'LE_fnum',
    'LE_fqc',
    'LE_fsd',
    'LE_fwin',
    'R_ref_CUT_REF' 
]
```

```python
df = df.drop(dropcols, axis=1)
df
```

## Restrict time range

```python
keepyears = df.index.year >= 2013
df = df[keepyears].copy()
df
```

# **Save dataframe as parquet file**

```python
filepath = save_parquet(filename="CH-DAV_FP2022.5_1997-2022_ID20230206154316_30MIN", data=df)
```
