# Import packages

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme('poster')
```

# Load data

```python
src = r"../data/ch-lae-fp2021-(2004-2020)/Level-4QCFs_Dataset_DIIVE-20210608-141431_Original-30T.diive.csv"
date_parser = lambda x: dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')  # For DIIVE csv files    
df = pd.read_csv(src, header=[0], delimiter=',', date_parser=date_parser, na_values=-9999, index_col=0, parse_dates=[0], skiprows=[1])
```

```python
df.columns
```

# Air temperature

For colors: https://materialui.co/colors

```python
df[['Tair_f']].count()
```

## Daily and yearly averages

```python
ta_da = df[['Tair_f']].resample('D').mean()
ta_da
```

```python
ta_yr = df[['Tair_f']].resample('A').mean()
ta_yr
```

```python
fig = plt.figure(figsize=(30, 9))
ax = fig.add_subplot()
ax.plot_date(x=ta_da.index, y=ta_da, label="daily average", color="#B0BEC5")
ax.plot_date(x=ta_yr.index, y=ta_yr, label="yearly average", ms=10, color="#EC407A")
ax.axhline(0, color='black', lw=2)
ax.set_title("Air temperature")
ax.legend();
```

## Diel cycles

```python
ta_hh = df[['Tair_f']].copy()

# Calculate the hourly mean per month
ta_means = ta_hh.groupby([ta_hh.index.month, ta_hh.index.time]).mean().unstack().T.droplevel(0, axis=0)
```

```python
# Plot
ta_means.plot(figsize=(20, 9), ls='-', cmap='jet')
plt.title("TA per hour and month")
plt.axhline(0, c='black', lw=1)
plt.legend();
```

## First 8 vs last 8
Compare the **first 8 years (2005-2012)** to the **most recent 8 years (2013-2020)**

```python
ta_hh = df[['Tair_f']].copy()

# First 8
ix_first8 = (ta_hh.index.year >= 2005) & (ta_hh.index.year <= 2012)
first8 = ta_hh[ix_first8]
first8 = first8.groupby([first8.index.month, first8.index.time]).mean().unstack().T.droplevel(0, axis=0)

# Last 8
ix_last8 = (ta_hh.index.year >= 2013) & (ta_hh.index.year <= 2020)
last8 = ta_hh[ix_last8]
last8 = last8.groupby([last8.index.month, last8.index.time]).mean().unstack().T.droplevel(0, axis=0)

fig = plt.figure(figsize=(20, 15))
ax = fig.add_subplot()
first8.plot(ax=ax, ls='--', cmap='Paired')
last8.plot(ax=ax, ls='-', cmap='Paired');
```

```python
diff = last8 - first8
# fig = plt.figure(figsize=(16, 9))
# ax = fig.add_subplot()
diff.plot(figsize=(20, 10), cmap='Paired', title="TA diel cycles: difference last8 - first8");
```

## Half-hours > 0°C

```python
ta_hh = df[['Tair_f']].copy()
ta_hh = ta_hh.loc[ta_hh.index.year > 2004]
ix_hh0 = (ta_hh < 0)
sums = ix_hh0.resample('A').sum()
sums
```

```python
sums.plot.bar(figsize=(20, 10), title="Number of half-hours < 0°C");
```

# NEE

```python
subset = df[['NEE_f', 'Tair_f', 'GPP_f', 'Reco', 'Rg_f']].copy()
subset
```

```python
subset = subset.resample('M').mean()
subset
```

```python
subset.plot(title="Monthly means", figsize=(20, 10));
```

```python
subset['MONTH'] = subset.index.month
subset['YEAR'] = subset.index.year
```

```python
sns.set(rc={"figure.figsize":(18, 4)})
sns.lmplot(data=subset, x="Tair_f", y="NEE_f", hue="MONTH", height=9, aspect=1.5).set(title='NEE');
plt.axhline(0, color='black');
```

```python
sns.lmplot(data=subset, x="Tair_f", y="GPP_f", hue="MONTH", height=9, aspect=1.5).set(title='GPP');
```

```python
sns.lmplot(data=subset, x="Tair_f", y="Reco", hue="MONTH", height=9, aspect=1.5).set(title='RECO');
```

```python
classes = 10
_plot_df = subset[['NEE_f', 'Tair_f', 'Rg_f']].copy()
_plot_df['TA_group'] = pd.qcut(_plot_df['Tair_f'], q=classes, labels=False)
_plot_df['SW_IN_group'] = pd.qcut(_plot_df['Rg_f'], q=classes, labels=False)
_plot_df = _plot_df.drop(['Tair_f', 'Rg_f'], axis=1)  # Remove data cols (index cols are kept)
_plot_df = _plot_df.groupby(['TA_group', 'SW_IN_group']).mean().unstack().droplevel(0, axis=1)
_plot_df
```

```python
f, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(_plot_df, annot=True, linewidths=.5, ax=ax, cmap='Spectral_r', fmt='.1f').set(title='NEE in 10 classes of SW_IN and TA');
```

```python
classes = 10
_plot_df = subset[['NEE_f', 'GPP_f', 'Reco']].copy()
_plot_df['GPP_group'] = pd.qcut(_plot_df['GPP_f'], q=classes, labels=False)
_plot_df['RECO_group'] = pd.qcut(_plot_df['Reco'], q=classes, labels=False)
_plot_df = _plot_df.drop(['GPP_f', 'Reco'], axis=1)  # Remove data cols (index cols are kept)
_plot_df = _plot_df.groupby(['RECO_group', 'GPP_group']).mean().unstack().droplevel(0, axis=1)
f, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(_plot_df, annot=True, linewidths=.5, ax=ax, cmap='Spectral_r', fmt='.1f').set(title='NEE in 10 classes of GPP and RECO');
```
