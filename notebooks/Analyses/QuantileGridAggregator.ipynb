# **Description**

- Calculates aggregated values of a target variable (z) within a 2D grid defined by quantiles of two independent variables (x and y).
- This example shows mean vapor pressure deficit (z) in classes of radiation (x) and air temperature (y) quantiles.
- See docstring below for additional details.

</br>

# **Load example data**

````python
df = dv.load_exampledata_parquet()
df
````

## Make subset

````python
# Make subset of three required columns
vpd_col = 'VPD_f'  # Vapor pressure deficit
ta_col = 'Tair_f'  # Air temperature
swin_col = 'Rg_f'  # Shortwave incoming radiation
subset = df[[vpd_col, ta_col, swin_col]].copy()
subset = subset.loc[(subset.index.month >= 5) & (subset.index.month <= 9)].copy()  # Use data May and Sep
daytime_locs = (subset[swin_col] > 0)  # Use daytime data 
subset = subset[daytime_locs].copy()
subset = subset.dropna()
````

````python
subset
````

</br>

# (1) **Calculate mean of z in 10 quantile classes of x and y**

````python
q = dv.qga(
    x=subset[swin_col],
    y=subset[ta_col],
    z=subset[vpd_col],
    n_quantiles=10,
    min_n_vals_per_bin=3,
    binagg_z='mean'
)
q.run()
````

````python
df_wide = q.df_wide.copy()
df_wide
````

````python
df_long = q.df_long.copy()
df_long
````

</br>

## Plot

````python
hm = dv.heatmapxyz(
    x=df_long['BIN_Rg_f'],
    y=df_long['BIN_Tair_f'],
    z=df_long['VPD_f'],
    cb_digits_after_comma=0,
    xlabel=r'Short-wave incoming radiation ($\mathrm{percentile}$)',
    ylabel=r'Air temperature ($\mathrm{percentile}$)',
    zlabel=r'Vapor pressure deficit (mean)'
)
hm.plot()
````

</br>

# (2) **Calculate counts of z in 100 quantile classes of x and y**

````python
q = dv.qga(
    x=subset[swin_col],
    y=subset[ta_col],
    z=subset[vpd_col],
    n_quantiles=100,
    min_n_vals_per_bin=3,
    binagg_z='count'
)
q.run()
````

````python
df_wide = q.df_wide.copy()
df_wide
````

````python
df_long = q.df_long.copy()
df_long
````

</br>

## Plot
- In this case there are many classes (100*100), therefore it is recommended to manually define `tickpos` and `ticklabels`.

````python
hm = dv.heatmapxyz(
    x=df_long['BIN_Rg_f'],
    y=df_long['BIN_Tair_f'],
    z=df_long['VPD_f'],
    cb_digits_after_comma=0,
    xlabel=r'Short-wave incoming radiation ($\mathrm{percentile}$)',
    ylabel=r'Air temperature ($\mathrm{percentile}$)',
    zlabel=r'Vapor pressure deficit (counts)',
    xtickpos=[25, 50, 75],
    xticklabels=['25', '50', '75'],
    ytickpos=[25, 50, 75],
    yticklabels=['25', '50', '75']
)
hm.plot()
````
