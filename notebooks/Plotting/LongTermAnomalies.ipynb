# **Load example data**

```python
df = pd.read_csv('data.txt', 
                 header=0, 
                 encoding='utf-8', 
                 delimiter=';',                 
                 index_col='time', 
                 dtype=None,
                 engine='python')
series = df['tre200y0'].copy()
series_label = "Air temperature"
series
```
