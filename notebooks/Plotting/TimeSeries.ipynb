# **Imports**

```python
import pkg_resources
from datetime import datetime
from diive.core.plotting.timeseries import TimeSeries
from diive.configs.exampledata import load_exampledata_pickle

dt_string = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print(f"This page was last modified on: {dt_string}")
version_diive = pkg_resources.get_distribution("diive").version
print(f"diive version: v{version_diive}")
```

# **Load data**

```python
df = load_exampledata_pickle()
df
```

```python
series = df['NEE_CUT_REF_orig'].copy()
series
```

# **Plot**

## Interactive plot for notebooks

```python
TimeSeries(series=series).plot_interactive()
```

## Static plot (non-interactive)

```python
TimeSeries(series=series).plot()
```
