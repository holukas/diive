# **Description**

- Make a ridge plot from time series data
- This example uses half-hourly air temperature data from 2019

</br>

# **Imports**

```python
import importlib.metadata
from datetime import datetime
import warnings
warnings.simplefilter("ignore", UserWarning)
import diive as dv
from diive.configs.exampledata import load_exampledata_parquet

version_diive = importlib.metadata.version("diive")
print(f"diive version: v{version_diive}")
```

</br>

# **Load example data**

```python
df = load_exampledata_parquet()
ta = df['Tair_f'].copy()
ta
```

</br>

# **Ridge plot per month**

## Example data for one year (2019)

```python
series = ta.loc[ta.index.year == 2019].copy()
series
```

## Make plot

```python
rp = dv.ridgeplot(series=series)  # Init RidgePlotTS instance
rp.per_month()
```

</br>

# **Ridge plot per month** with full parameters

## Example data for one year (2019)

```python
series = ta.loc[ta.index.year == 2019].copy()
series
```

## Make plot

```python
rp = dv.ridgeplot(series=series)
rp.per_month(
    xlim=None,  # min/max as list
    ylim=None,  # min/max as list
    hspace=-0.5,  # overlap between months
    xlabel=r'Air temperature (Â°C)',    
    fig_width=8,
    fig_height=10,
    shade_percentile= 0.5,
    show_mean_line=False,
    fig_title= "Air temperatures per month (2019)",
    fig_dpi=72,
    showplot=True,
    ascending=False
)
```
